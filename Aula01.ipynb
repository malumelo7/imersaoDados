### Desafio 6. Renomear as colunas tirando o hífen;

```python
novo_nome_coluna = []
nome_das_colunas = dados.columns
for coluna in nome_das_colunas:
  coluna = coluna.replace('-','_')
  novo_nome_coluna.append(coluna)
dados.columns = novo_nome_coluna

dados.head()
```

### Desafio 7. Deixar os gráficos com título, label (Matplotlib, pyplot);

```python
dados['tratamento'].value_counts().plot.bar()

import numpy as np
import matplotlib.pyplot as plt

plt.title("Quantitativo de Tratamento")
plt.xlabel("Tipo de Tratamento")
plt.ylabel("Quantidade")
```

```python
dados_copy = dados.copy()

graphic = dados_copy['tratamento'].value_counts().plot.bar()

import numpy as np
import matplotlib.pyplot as plt

plt.title("Quantitativo de Tratamento")
plt.xlabel("Tipo de Tratamento")
plt.ylabel("Quantidade")
plt.bar(dados['tratamento'].unique(),dados['tratamento'].value_counts() ,color = ["blue", "red"])
graphic.set_xticklabels(labels = ['Com Droga', 'Com Controle'])
```
